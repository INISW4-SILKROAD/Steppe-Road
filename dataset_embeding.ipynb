```python
import torch
from imagebind.data import load_and_transform_vision_data
import json
import os
from tqdm import tqdm
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
with open('dataset/dataset.json', 'rb') as f:
    all_ = json.load(f)
```

```python
dic = {}
for i in tqdm(all_):
    dic[i['id']] = load_and_transform_vision_data(i['img'], device).to('cpu')
```

```python
dic2 = {}
for item in all_:
    for index, name in enumerate(item['img']):
        dic2[name] = dic[item['id']][index]
    
```
